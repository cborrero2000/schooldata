```python
orders = pd.read_csv("http://bit.ly/chiporders", sep='\t')
orders
```

```python
orders.head(6)
```

```python
orders.item_price = orders.item_price.str.replace('$', '').astype(float)
#orders.item_price.str.replace('$', '').astype(float).sum()
```

```python
orders[orders.order_id==1].item_price.str.replace('$','').astype(float).sum()
```

```python
orders.groupby('order_id').item_price.sum()
orders.groupby('order_id').item_price.agg(['sum', 'count', 'mean'])
```

```python
orders['total'] = orders.groupby('order_id').item_price.transform(sum)
```

```python
orders
```

```python
orders['percent_of_total'] = orders.item_price/orders.total
orders
```

```python
ted = pd.read_csv('ted.csv')
ted.head(2)
```

```python
ted['views_per_comments'] = ted.views / ted.comments
ted.sort_values('views_per_comments').head()
```

```python
%matplotlib inline
ted[ted.comments < 1000].comments.plot(kind='hist')
```

```python
ted.query('comments<1000').comments.plot(kind='hist')
```

```python
ted.loc[ted.comments < 1000, 'comments'].plot(kind='hist', bins=20)
```
